```python
import re;

fetch = requests.urlopen("https://sherlock-holm.es/stories/plain-text/case.txt");

#help(fetch);


words = str(fetch.read().decode('utf8')).split(' ');

word_counts = {};

for wordString in words:
    # use regex to strip non-alphanumeric characters from string 
    # such as periods and other punctuation
    word = re.sub(r'\W+', '', wordString.lower());
    
    # if the word is made up of only aphabetical characters,
    # initialize an entry in the word_count dictionary if none exists else increment it by one
    if(word.isalpha()):
        try:   
            word_counts[word]+=1;
        except:
            word_counts[word]=1;

print(word_counts);
```
